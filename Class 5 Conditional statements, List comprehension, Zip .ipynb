```python
# loops
# for, while
# controlling loops using break, continue and pass statements

numbers = [1,3,5,7,9,11,13]
for n in numbers:
    print(n)
    if n >= 10:
        break
```

```python
numbers = [1,3,5,7,9,11,13]
for n in numbers:
    if n >= 10:
        print(n+1)
        break
```

```python
numbers = [1,3,5,7,9,11,13]
for n in numbers:
    if n >= 5:
        print(n+1)
        continue
```

```python
upper_num = 10
count = 0
while count < upper_num:
    print(count)
    if count >= 3:
        break
    count += 1
```

```python
upper_num = 10
count = 0
while count < upper_num:
    print(count)
    if count >= 3:
        pass
    count += 1
```

```python
upper_num = 10
count = 0
while count < upper_num:
    print(count)
    if count >= 3:
        continue
        count += 1
```

```python
# enumerate.
s = 'Hello'
for letter in s:
    print(s.index(letter), letter)
```

```python
s = 'Hello'
for i, letter in enumerate(s):
    print(i, letter)
```

```python
s = 'Hello'
for i, letter in enumerate(s, start = 1):
    print(i, letter)
```

```python
numbers = list(range(10))
# create a new list as the square of each number in the numbers list.
nl = [ num ** 2 for num in numbers ]
nl
```

```python
nl = [ num ** 2 for num in numbers if num % 2 == 0 ]
nl
```

```python
nl = []
for num in numbers:
    if num % 2 == 0:
        nl.append(num** 2)
nl
```
