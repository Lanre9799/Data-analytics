```python
filt_df = (df['Country'] != 'United Kingdom')
filt_df = df[filt_df]
grouped_df = filt_df.groupby('Country')['SaleRevenue'].sum()
sorted_df = grouped_df.sort_values(ascending=False)
sorted_df
```

```python
sorted_df.plot(kind= 'bar' , x= 'Country', y='Value', legend=False)
plt.title('Sales Revenue by Country, Excluding UK')
plt.xlabel('Countrie')
plt.ylabel('Sales Revenue')
plt.grid(True)
plt.show()
```

```python
sorted_df.plot(kind= 'barh' , x= 'Country', y='Value', legend=False)
plt.title('Sales Revenue by Country, Excluding UK')
plt.xlabel('Countries')
plt.ylabel('Sales Revenue')
plt.grid(True)
plt.show()
```

```python
cust_df = df.groupby('CustomerID')['SaleRevenue'].sum()
cust_df = cust_df.nlargest(10)
cust_df
```

```python
cust_df.plot(kind= 'bar' , x= 'Country', y='Value', legend=False)
plt.title('Top 10 Customers by Sales Revenue')
plt.xlabel('Customers')
plt.ylabel('Sales Revenue (GBP)')
plt.grid(True)
plt.show()
```
