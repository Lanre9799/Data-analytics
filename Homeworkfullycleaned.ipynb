```python


prod_df.plot(kind= 'bar' , x= 'Country', y='Value', legend=False)
plt.title('distribution of happiness rank in each country')
plt.xlabel('Country')
plt.ylabel('Happiness score')
plt.grid(True)
plt.show()
```
