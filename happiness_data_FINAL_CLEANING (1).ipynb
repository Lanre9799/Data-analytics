```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"happiness_data_UPDATED.csv")
df.head()
```

```python
filt = df['Country'] == "Nigeria"
df[filt]
```

```python
df.info()
```

```python
# get the unique list of countries in the dataset.
unique_countries = list(df['Country'].unique())
unique_countries.sort()
for country in unique_countries:
    for rowid in df.index:
        if df.loc[rowid, 'Country'] == country:
            # country match found.
            # create a filter getting all data pertaining to the currently selected country
            filt = (df['Country'] == country)
            # get the region from columns that have region provided
            for row in df[filt].index:
                if isinstance(df.loc[row, 'Region'], str):
                    region = df.loc[row, 'Region']
                    break
            # replace the NaN with the correct region where region is missing.
            for row in df[filt].index:
                if isinstance(df.loc[row, 'Region'], float):
                    df.loc[row, 'Region'] = region
df.info()
```

```python
df.head()
```

```python
# get the each country name, manually check for spellings. 
unique_countries = list(df['Country'].unique())
unique_countries.sort()
for country in unique_countries:
    if country[0].upper() == 'T':
        print(country)
```

```python
# replace country names that were misspelled.
def update_countries(country):
    if country == 'Hong Kong S.A.R., China':
        return 'Hong Kong'
    elif country == 'Northern Cyprus':
        return 'North Cyprus'
    elif country == 'Somaliland region':
        return 'Somaliland Region'  
    elif country == 'Trinidad & Tobago':
        return 'Trinidad and Tobago'
    elif country == 'Taiwan Province of China':
        return 'Taiwan' 
    else:
        return country

df['Country'] = df['Country'].map(update_countries)
df.head(50)
```
