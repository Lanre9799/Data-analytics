```python
#  updating multiple rows in a column.
# this can be done using functions such as
# apply, map. Others are apply map.

# apply function updates all the row values in a column
# map function updates selected values.

# apply works with both series and dataframes.
# we mostly use it with series.

# let's say we want to add 1 to every student's score.

def update_score(score):
    return score + 1

df['student_scores'] = df['student_scores'].apply(update_score)
df
```

```python
# rewrite the above using a lambda expression
# the lambda function will deduct 1 from all the scores

df['student_scores'] = df['student_scores'].apply(lambda score: score - 1)
df
```

```python
df.drop('student_score', axis=1, inplace= True)
```

```python
df
```

```python
# map method only works with series data
# let's say we only want to add 5 to students that scored below 50
# and also adjust their grade to "C"

def update_student_score(score):
    if score < 50:
        return score + 5
    else:
        return score

df['student_scores'] = df['student_scores'].map(update_student_score)
df
```

```python
# update the grades column as follows.
# if score is above 70, grade is 'A'
# score 60 - 69, grade is 'B'
# score 50 - 59, grade is 'C'
# score 45 - 49, grade is 'D'
# score 40 - 44, grade is 'E'
# score below 40, grade is 'F'

df.index
```

```python
for d in df.index:
    print(d)
```

```python
# consider the codes for updating 1 or more columns of a row of data.
def update_grades(score):
    if score >= 70:
        df.loc[:,'grades'] = 'A'
    elif 69 >= score >= 60:
        df.loc[:,'grades'] = 'B'
    elif 59 >= score >= 50:
        df.loc[:,'grades'] = 'C'
    elif 49 >= score >= 45:
        df.loc[:,'grades'] = 'D'
    elif 44 >= score >= 40:
        df.loc[:,'grades'] = 'E'
    else:
        df.loc[:,'grades'] = 'F'
        return score
        

df['student_scores'] = df['student_scores'].map(update_grades)
df
```

```python
df.loc[:, ['grades']] = df.loc[:, ['grades']].apply(lambda x: [x[0] * 2, x[1] + 1])
```
